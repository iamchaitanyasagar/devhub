```python
df = pd.read_csv('housing.csv')
df.head()
```

```python
print('Missing Values: ')
print(df.isnull().sum())
print('\n')
print('Data Types: ')
print(df.dtypes)
```

```python
df.dropna(inplace = True)
```

```python
df['ocean_proximity'].value_counts()
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
le.fit(df["ocean_proximity"])
df["ocean_proximity"] = le.transform(df["ocean_proximity"])
```

```python
df
```

```python
# # Dropping categorical variable and missing values
# df.drop('ocean_proximity', axis = 1, inplace = True)
# df.dropna(inplace = True)
```

```python
from sklearn.linear_model import LinearRegression # Importing Linear Regression Model
from sklearn.model_selection import train_test_split # Importing Model Selection to split dataset into training and testing set
from sklearn.metrics import r2_score, mean_squared_error # Importing R² score
from math import sqrt
```

```python
X = df.drop('median_house_value', axis = 1)# Selecting independent variables by removing 'median_house_value' column from dataframe
y = df['median_house_value'] # Selecting dependent variable from the dataframe
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, # Splitting X and y instances for training and testing sets
                                                   test_size = 0.2, random_state = 0) # 20% of data will be used for testing the model
```

```python
lr = LinearRegression() # Initializing the model
lr.fit(X_train, y_train) # Fitting onto the training data

y_pred = lr.predict(X_test) # Predicting prices on the testing set
score = r2_score(y_test, y_pred) # Measuring Model performance by comparing the predicted values to the actual values

print(f'\nModel R² Score: {score:.3f}') # Printing R² score
print("Mean Squared Error :", sqrt(mean_squared_error(y_test, y_pred)))
```

```python
import plotly.express as px
import plotly.graph_objs as go
from math import sqrt

data = pd.DataFrame({'y_test':y_test, 'y_pred':y_pred})
fig = px.scatter(data, x='y_test', y='y_pred', trendline='ols', trendline_color_override='red')

fig.update_layout(
    title='R² = %0.3f' % score,
    xaxis_title='y_test',
    yaxis_title='y_pred',
    template = 'ggplot2',
    height = 200
)

trendline_results = px.get_trendline_results(fig)
r_squared = trendline_results.iloc[0]['px_fit_results'].rsquared

fig.update_layout(height=1000)
fig.update_xaxes(showgrid=True)
fig.update_yaxes(showgrid=True)
fig.show()
```

Decision Tree Regressor

```python
from sklearn.tree import DecisionTreeRegressor
dt_r = DecisionTreeRegressor()

dt_r.fit(X_train, y_train) # Fitting onto the training data

y_pred = dt_r.predict(X_test) # Predicting prices on the testing set
score = r2_score(y_test, y_pred) # Measuring Model performance by comparing the predicted values to the actual values

print(f'\nModel R² Score: {score:.3f}') # Printing R² score
print("Mean Squared Error :", sqrt(mean_squared_error(y_test, y_pred)))
```

```python
# KNN
```

```python
from sklearn.neighbors import KNeighborsRegressor

for i in range(10,20):
  print(i)
  neigh = KNeighborsRegressor(n_neighbors=i)
  neigh.fit(X_train, y_train) # Fitting onto the training data

  y_pred = neigh.predict(X_test) # Predicting prices on the testing set
  score = r2_score(y_test, y_pred) # Measuring Model performance by comparing the predicted values to the actual values

  print(f'\nModel R² Score: {score:.3f}') # Printing R² score
```

```python
neigh = KNeighborsRegressor(n_neighbors=18)
neigh.fit(X_train, y_train) # Fitting onto the training data

y_pred = neigh.predict(X_test) # Predicting prices on the testing set
score = r2_score(y_test, y_pred) # Measuring Model performance by comparing the predicted values to the actual values

print(f'\nModel R² Score: {score:.3f}') # Printing R² score
print("Mean Squared Error :", sqrt(mean_squared_error(y_test, y_pred)))
```

```python
import xgboost as xgb
xg_reg = xgb.XGBRegressor(objective ='reg:linear', colsample_bytree = 1,eta=0.3, learning_rate = 0.1,
                max_depth = 5, alpha = 10, n_estimators = 2000)
xg_reg.fit(X_train, y_train)

y_pred = xg_reg.predict(X_test)

score=r2_score(y_test,y_pred)
print('the R squared of the xgboost method is:', score)
```
